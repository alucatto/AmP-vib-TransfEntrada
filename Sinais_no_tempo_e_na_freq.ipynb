# Exercício didático sobre processamento digital de sinais
Demonstração de sinais no domínio do tempo e no domínio da frequência

Bibliotecas

```
import math
import numpy as np
import matplotlib.pyplot as plt
```

Parâmetros do sinal

```
#Parâmetros
f = 10 #Frequencia
fs = 200 #Freq amostragem
D = 1 #Duração da janela

#Bases
Ts = 1/fs #Periodo de amostragem
N = D*fs #Número de pontos

N_vis=int(N/2) #Visualização do espectro

x = np.linspace(0, D, N)
```

Sinal senoidal

```
y1 = np.sin(2*np.pi*f*x) #sinal senoidal

plt.plot(x,y1)
plt.xlabel('Tempo (s)')
```

Sinal quadrado

```
y2=np.zeros(N)

#sinal quadrado
for i in range(0,N):
  if y1[i] > 0.5: 
    y2[i] = 1
  else:
    y2[i] = 0

plt.plot(x,y2)
plt.xlabel('Tempo (s)')
```

Ambos os sinais juntos, para visualização

```
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('Tempo (s)')
```

Espectro do sinal senoidal

```
Y1 = 2*np.abs(np.fft.fft(y1)/np.shape(y1)[0])

plt.plot(Y1[:N_vis])
plt.grid()
plt.xlabel('Frequência (Hz)')
```

Espectro do sinal quadrado

```
Y2 = 2*np.abs(np.fft.fft(y2)/np.shape(y2)[0])

plt.plot(Y2[:N_vis])
plt.grid()
plt.xlabel('Frequência (Hz)')
```
